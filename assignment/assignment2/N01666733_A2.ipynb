```python
from PIL import Image, ImageDraw, ImageFilter, ImageFont
from IPython.display import display
import pytesseract
```

```python
img = Image.open("pagani-red.jpeg")
img.show()
```

```python
draw = ImageDraw.Draw(img)
text = "Itopa Asher Anumah"
font_size = 80
font_path = 'ButterflyKids-Regular.ttf'
font = ImageFont.truetype(font_path, font_size)

draw.text((80, 50), text, fill=(255, 0, 255),font=font)  # Adjust position and color as needed
img.show()
```

```python
img = img.convert("RGB")
d = img.getdata()
new_image = []
for item in d:
    
    if item[0] > 150 and item[1] < 100 and item[2] < 100:
        new_image.append((0, 255, 0))
    else:
        new_image.append(item)
img.putdata(new_image)
display(img)
```

```python
text = pytesseract.image_to_string(img, lang="eng", config='--psm 12')
print("Extracted Text: ", text)
```

```python
edges = img.convert("L").filter(ImageFilter.FIND_EDGES)
edges.show()
```
